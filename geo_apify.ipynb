```python
import pandas as pd
import numpy as np
import requests 
import json
apiKey='0ce9de199a3449a8a43b70421678e0ae'
base_url = 'https://api.geoapify.com/v1/geocode/search?'
```

```python
city_housing_df = pd.read_csv('./Resources/housing_city.csv')
```

```python
city_housing_df.dropna(inplace=True)
city_housing_df
```

```python
state_housing_df = pd.read_csv('./Resources/housing_state.csv')
state_housing_df
```

```python
combined_df = state_housing_df.copy()

lat = []
lon = []
for state in combined_df['state']:
    url = base_url+'text='+state+'&lang=en&limit=10&type=state&filter=countrycode:us&&format=json&apiKey='+apiKey
    response = requests.get(url).json()
    lat.append(response['results'][0]['lat'])
    lon.append(response['results'][0]['lon'])
combined_df['lat'] = lat 
combined_df['lon'] = lon  
combined_df.to_csv("location_housing_state.csv",index=False)
combined_df
```

```python
combined_df2 = city_housing_df.copy()

missing = []

lat = []
lon = []
for city in combined_df2['city']:
    url = base_url+'text='+city+'&lang=en&limit=10&type=city&filter=countrycode:us&&format=json&apiKey='+apiKey
    response = requests.get(url).json()
    if response['results']:
        lat.append(response['results'][0]['lat'])
        lon.append(response['results'][0]['lon'])
    else:
        missing.append('city')  
        lat.append(np.nan)
        lon.append(np.nan)
        
combined_df2['lat'] = lat 
combined_df2['lon'] = lon  
combined_df2.to_csv("location_housing_city.csv",index=False)
combined_df2
```

```python
location_housing_state_df = pd.read_csv('location_housing_state.csv')
location_housing_state_df
```

```python
with open('us-states.json', 'r') as file:
    existing_data = json.load(file)
```

```python


# Modify the existing data by adding the "price" key to the properties dictionary for each feature
for feature in existing_data["features"]:
    state = feature['properties']['name']
    if state in location_housing_state_df['state'].values:
        feature["properties"]["price"] = location_housing_state_df[location_housing_state_df['state']==state]['price'].values[0]  # Set the price value here
    else:
        existing_data['features'].remove(feature)
       
    

# Write the modified data back to the JSON file
with open('us_state_prices.geojson', 'w') as file:
    json.dump(existing_data, file, indent=0)
```
